```python
## import libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
### plot the fundamental diagram
k0 = 5
u = 1.5
tf = 1.0
Qk = lambda k: k/tf*np.exp(-(k/k0)**u/u)
kar = np.linspace(0.01,k0*1.2,20)
plt.plot(kar,Qk(kar))
plt.xlabel("k")
plt.ylabel("q")
```

```python
## plot the fundamental diagram as a q/t relation
Tk = lambda k: k/Qk(k)
plt.plot(Qk(kar),Tk(kar))
plt.xlabel("q")
plt.ylabel("t")
```

```python
### demand
a = 4
b= 1
Qd = lambda t: a - b*t
tar = Tk(kar)
plt.plot(Qd(tar),tar)

### plot supply alongside demand
plt.plot(Qk(kar),tar)
plt.xlabel("q")
plt.ylabel("t")
```

```python
### find the intersectioin
from scipy.optimize import root
kIntersection = root(lambda k: Qk(k) - Qd(Tk(k)),1.0).x[0]
Qk(kIntersection)
```
